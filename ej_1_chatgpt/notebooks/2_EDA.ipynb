```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos
df = pd.read_parquet('../data/df_1.parquet')
df.head()
```

```python
df.info()
```

```python
# Gráfico de series de tiempo para ventas y precios

plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(df['date'], df['demand'], label = 'Demand')
plt.title('Sales Demand Over Time')

plt.subplot(2,1,2)
plt.plot(df['date'], df['price'], label='Price', color='orange')
plt.title('Product Price Over Time')
plt.tight_layout()
plt.show()
```

```python
# Ver la correlación entre las variables

sns.heatmap(df[['price', 'demand', 'promo']].corr(), annot=True, cmap='coolwarm')
plt.show()
```
