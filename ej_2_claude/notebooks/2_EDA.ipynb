```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos

df = pd.read_parquet('../data/df_1.parquet')
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(211)
plt.plot(df['fecha'], df['precio'])
plt.title('Precio a lo largo del tiempo')
plt.subplot(212)
plt.plot(df['fecha'], df['demanda'])
plt.title('Demanda a lo largo del tiempo')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['precio'], df['demanda'])
plt.xlabel('Precio')
plt.ylabel('Demanda')
plt.title('Relacin Precio-Demanda')
plt.show()
```

```python
# Ver la correlacin entre las variables

sns.heatmap(df[['precio', 'demanda']].corr(), annot=True, cmap='coolwarm')
plt.show()
```
